```python
def avg_number(*args):
  result = 0
  for i in args:
    result += i
  avg = result / len(args)
  return avg
avg_number(1,2,3)
```

```python
# 1개의 스레드 

import time

def long_task(): # 3초의 시간이 걸리는 함수 
  for i in range(3):
    time.sleep(1) # 1초대기
    print("now loading... %s\n" %i)
print('start')


for i in range(3): #now loading이 3번 반복됨
  long_task()

print("Hello Player")
```

```python
import time
import threading

def long_task(): #5초의 시간이 걸리는 함수
    for i in range(5):
        time.sleep(1) # 1초 대기
        print("%s입금 시 사기전화를 조심하세요!\n"% i)
print('start')

threads = []
for i in range(5):
    t = threading.Thread(target = long_task)
    threads.append(t)
for t in threads:
    t.start()
for t in threads:
    t.join() #join으로 스레드가 종료할 때 까지 기다린다
print('입금되었습니다')
```
