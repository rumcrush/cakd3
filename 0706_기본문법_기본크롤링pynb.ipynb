```python
coffee = 5

while 1:
    money = int(input("300원으로 한정 아메리카노를 즐겨보세요"))
    if money == 300:
        print('맛있게 드세요')
        coffee -= 1
    elif money > 300:
        print('맛있게 드시고 거스름돈 받아가세요')
    elif money < 300:
        print('저런! 금액이 부족합니다!')
    if coffee == 0:
        print('한정 판매가 끝났습니다')
        
        break
```
