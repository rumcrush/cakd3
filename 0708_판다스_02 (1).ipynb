```python
#데이터 프레임 생성
import numpy as np
import pandas as pd

np.random.seed(0)
data = np.random.randint(80,100,size=(5,5))
df = pd.DataFrame(data, columns=list('abcde'))
df
```

```python
df1 = df.copy()
df1
```

```python
#숫자 인덱스가 아닌 다른 인덱스 넣기
df1['index'] = list('가나다라마')
df1
```

```python
#인덱스 위치 바꾸기 
df2 = df1.set_index('index')
df2
```

```python
#원래 인덱스로 돌리기 #원래 인덱스에 있던 값은 컬럼으로 들어감
df3 = df2.reset_index('index')
df3
```

```python
#컬럼의 순서 바꾸기  #컬럼을 리스트 안에 넣어서 재배열
df4 = df3[['a','b','c','d','e','index']] #위치순서대로 컬럼 이름을 넣어준다
df4
```

```python
#숫자 인덱스 추가로 넣기
df1['f'] = [80,81,82,83,84]
df1
```

```python
#Q.딕셔너리로 5행 5열 데이터프레임을 작성 

index = ['r0','r1','r2','r3','r4']
columns = ['abcde']

import pandas as pd

dict_data = {'a' : [1,2,3,4,5], 'b':[4,5,6,7,8], 'c' : [7,8,9,10,11], 'c':[12,13,14,15,16],'e':[20,21,22,23,24]}
df = pd.DataFrame(dict_data, index=['r0','r1','r2','r3','r4'])

df

```

```python
df1.loc['r5'] = 0
df1.loc['r6'] = 0
```

```python
#새로운 인덱스 추가, reindex(new_index,fill_value=0)
new_index = ['r0','r1','r2','r3','r4','r5','r6','r7','r8']
df5 = df.reindex(new_index, fill_value=0)
df5
```

```python
#r5-r8까지 인덱스 삭제 후 출력 
df5 = df5.drop(['r5','r6','r7','r8'],axis=0)
df5
```

```python
#인덱스 내림차순으로 정렬 
df1_s = df5.sort_index() #인덱스 오름차순 정렬 (sort_index())
df1_s = df5.sort_index(ascending=False)
df1_s
```

```python
#c라는 컬럼을 기준으로 오름차순 정렬
df1_c = df1.sort_values(by='c')
df1_c
```

```python
#c라는 컬럼을 기준으로 내림차순 정렬
df1_c = df1.sort_values(by='c', ascending=False)
df1_c
```

```python
#행과 열이 바뀌는 전치
df1_t = df1.transpose() #transpopse()를 쓴다
print(df1_t)

print()

df2_t = df1_t.transpose()
print(df2_t)
```
